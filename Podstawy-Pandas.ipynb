# Tworzenie dodatkowych kolumn z użyciem logiki

```python
df['PURCHASE_EQUAL_TO_LIMIT'] = np.where( df.PURCHASES == df.CREDIT_LIMIT, True, False)
```

```python
cond = [(df.PURCHASES >= df.CREDIT_LIMIT) & (df.CREDIT_LIMIT >= 45000),
        (df.PURCHASES >= df.CREDIT_LIMIT) & (df.CREDIT_LIMIT < 45000),
        (df.PURCHASES < df.CREDIT_LIMIT) & (df.CREDIT_LIMIT >= 45000),
        (df.PURCHASES < df.CREDIT_LIMIT) & (df.CREDIT_LIMIT < 45000)
        ]

choice = ['S1', 'S2', 'S3', 'S4']

df['SCENARIOES'] = np.select( cond, choice, default= 'S5')
```

# GroupBy i Pivot Table

```python
df.groupby(['CUST_ID'])[['BALANCE','PURCHASES']].sum()
```

```python
df.groupby(['CUST_ID'], as_index = False).agg({'BALANCE' : 'sum', 'PURCHASES' : 'sum', 'CREDIT_LIMIT' : 'mean'})
```

```python
pivot = df.pivot_table(index= 'SCENARIOES',
               columns=['PURCHASE_EQUAL_TO_LIMIT'],
               values=['BALANCE'],
               aggfunc=['sum'],
               fill_value = 0)\
               .join(df.pivot_table(index= 'SCENARIOES',
               columns=['PURCHASE_EQUAL_TO_LIMIT'],
               values=['PURCHASES'],
               aggfunc=['sum'],
               fill_value = 0))\
               .join(df.pivot_table(index= 'SCENARIOES',
               columns=['PURCHASE_EQUAL_TO_LIMIT'],
               values=['CREDIT_LIMIT'],
               aggfunc=['mean'],
               fill_value = 0)).fillna(0)
pivot
```

```python
pivot.swaplevel(axis=1).sort_index(axis=1, level=1)
```

# Łączenie DataFrame'ów
